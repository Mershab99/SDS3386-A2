# Question 1 (2 points)

Using the basketball data in "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv", determine the proportion of Home games ('game_location' = 'H') won, and the proportion of Away games won, over all the seasons of the NBA recorded in this dataset.

*Hint: run 'df.value_counts?'*

```python
import pandas as pd  
import altair as alt
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/fivethirtyeight/data/master/nba-elo/nbaallelo.csv"
data = pd.read_csv(url)
data
```

```python
#Filtering the data
won = data['game_result']== 'W'
home = data['game_location']=='H' 
away = data['game_location']=='A'

won_home = data[home & won]
won_away = data[away & won]

#Computing the proportions needed
won_home_proportion = won_home.shape[0]/data.shape[0]
won_away_proportion = won_away.shape[0]/data.shape[0]




print(f"Proportion of Home games won : {won_home_proportion} ({won_home_proportion*100:.2f} %)")

print(f"Proportion of Away games won : {won_away_proportion} ({won_away_proportion*100:.2f} %)")

```

# Answer:

Proportion of Home games won : 0.3112 (31.12 %)

Proportion of Away games won : 0.1887 (18.87 %)

The data is not tidy because:

# B)

```python
asst1 =grades['Asst1']
asst2 =grades['Asst2']
asst3 =grades['Asst3']
exam =grades['Exam']

final = asst1 + asst2 + asst3 + (70*exam/100)

final.loc[final.isnull()] = 'INC'
final

grades_calculated = grades.copy()
grades_calculated['Final'] = final
grades_calculated
```

# C

```python
#tidying the data

to_chart = grades_calculated.copy()
to_chart = to_chart.drop(index = 4)
to_chart.loc[to_chart['Asst2'].isnull(),'Asst2'] = 0
to_chart.loc[to_chart['Exam'].isnull(),'Exam'] = 0
to_chart.loc[to_chart['Final']== 'INC','Final'] = 0

#Clean dataframe to work with
to_chart
```

```python

a1_chart = alt.Chart(to_chart).mark_bar().encode(

    x = alt.X('Name', title =""),
    y = alt.Y('Asst1',scale=alt.Scale(domain=[0,10])),
    color = alt.Color('Name', scale=alt.Scale(scheme='pastel1'))
).properties(title='Assignment 1 grades')


a2_chart = alt.Chart(to_chart).mark_bar().encode(

     x = alt.X('Name', title =""),
    y = alt.Y('Asst2',scale=alt.Scale(domain=[0,10])),
    color = 'Name'
).properties(title='Assignment 2 grades')


a3_chart = alt.Chart(to_chart).mark_bar().encode(

     x = alt.X('Name', title =""),
    y = alt.Y('Asst3',scale=alt.Scale(domain=[0,10])),
   color = 'Name'
).properties(title='Assignment 3 grades')

exam_chart = alt.Chart(to_chart).mark_bar().encode(

    x = alt.X('Name', title =""),
    y = alt.Y('Exam',scale=alt.Scale(domain=[0,100])),
    color = 'Name'
).properties(title='Exam grades')

#TODO: Stacked bar chart of the student grades


a1_chart | a2_chart | a3_chart |exam_chart 
```
